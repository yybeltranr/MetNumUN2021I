## Primero selecciono los 6 mejores stocks de Yahoo Finance en los últimos 2 años y seis meses.

### *TESLA*

```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        self.SetStartDate(2018, 11, 4)
        self.SetEndDate(2021, 4, 4)

        #1,2. Select TSLA minute resolution data and set it to Raw normalization mode
        self.iwm = self.AddEquity("TSLA", Resolution.Minute)
        self.iwm.SetDataNormalizationMode(DataNormalizationMode.Raw)

    def OnData(self, data):

        #3. Place an order for 100 shares of TSLA and print the average fill price
        #4. Debug the AveragePrice of TSLA
        if not self.Portfolio.Invested:
            self.MarketOrder("TSLA", 100)
            self.Debug(str(self.Portfolio["TSLA"].AveragePrice))
```

## Luego reviso la compra de acciones para mantener el rendimiento en los últimos 2 años, iniciando con 1'000.000 de acciones y probando para 400 acciones.

```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        # 1. Set Starting Cash 
        self.SetCash(1000000)
        self.SetStartDate(2019, 4, 1)
        self.SetEndDate(2021, 4, 1)

        # Select stocks and add in DataNormalizationMode = Raw
        self.tsla = self.AddEquity("TSLA", Resolution.Minute)
        self.tsla.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.amzn = self.AddEquity("AMZN", Resolution.Minute)
        self.amzn.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.nflx = self.AddEquity("NFLX", Resolution.Minute)
        self.nflx.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.msft = self.AddEquity("MSFT", Resolution.Minute)
        self.msft.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.goog = self.AddEquity("GOOG", Resolution.Minute)
        self.goog.SetDataNormalizationMode(DataNormalizationMode.Raw)
        

    def OnData(self, data):

        if not self.Portfolio.Invested:
            self.MarketOrder("TSLA", 400)
            self.Debug(str(self.Portfolio["TSLA"].AveragePrice))
            self.MarketOrder("AMZN", 400)
            self.Debug(str(self.Portfolio["AMZN"].AveragePrice))
            self.MarketOrder("NFLX", 400)
            self.Debug(str(self.Portfolio["NFLX"].AveragePrice))
            self.MarketOrder("MSFT", 400)
            self.Debug(str(self.Portfolio["MSFT"].AveragePrice))
            self.MarketOrder("GOOG", 400)
            self.Debug(str(self.Portfolio["GOOG"].AveragePrice))
```
